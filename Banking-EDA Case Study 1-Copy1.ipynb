```python
categorical_cols = df[[
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())

    
```

```python


sns.set_theme(style="whitegrid")

for i, predictor in enumerate(df[[
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]].columns):

    plt.figure(figsize=(8, 5))
    
    sns.countplot(
        data=df,
        x=predictor,
        hue='Nationality',
        palette="Set2"  
    )
    
    plt.title(f"{predictor} vs Gender", fontsize=14, fontweight="bold")
    plt.xticks(rotation=0, ha="right")
    plt.tight_layout()
    plt.show()

  
```
